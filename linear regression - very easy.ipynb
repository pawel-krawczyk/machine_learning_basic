```
#import basic libraries

#data management
import numpy as np
import pandas as pd

#ML
from sklearn import linear_model

#visualisation
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#read data
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/homeprices.csv")
df
```

```
#using matplotlib.pyplot as plt create scattered plot
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price)
```

```
#create LinearRegression object
reg=linear_model.LinearRegression()
```

```
#train model
reg.fit(df[["area"]],df.price)
```

```
#predict sample value
reg.predict([[3300]])
```

```
# check coef -> a in ax+b
reg.coef_ 
```

```
# check intercept ->b in ax+b
reg.intercept_ 
```

```
#red now dataset
areas=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/areas.csv")
```

```
#predict prices
areas["prices"]=reg.predict(areas)

areas
```

```
#save to CSV
areas.to_csv("predictions.csv",index=False)
```

```
#visualise
plt.xlabel("area")
plt.ylabel("price")

plt.scatter(df.area,df.price,color="red", marker='+')
plt.plot(df.area, reg.predict(df[["area"]]))
```

## Second example - predicting canada gdp per capita in 2020

```
#load data
can=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv")

can.columns
```

```
#create object, train and predict
reg_can=linear_model.LinearRegression()
reg_can.fit(can[["year"]],can['per capita income (US$)'])
reg_can.predict([[2020]])
```

```
#visualise
plt.scatter(can.year, can['per capita income (US$)'])
plt.plot(can.year, reg_can.predict(can[['year']]), color="red")
```
