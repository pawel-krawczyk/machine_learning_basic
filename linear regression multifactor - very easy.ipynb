```
#import basic libraries

#data management
import numpy as np
import pandas as pd

#ML
from sklearn import linear_model

#visualisation
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#load data
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/homeprices.csv")
df
```

```
#import math library
import math

#get median
mb=math.floor(df.bedrooms.median())
mb
```

```
#populate nulls with median value
df.bedrooms.fillna(mb, inplace=True)
```

```
#display data
df
```

```
#create LinearRegression object
reg=linear_model.LinearRegression()
```

```
#train model with 3 variables
reg.fit(df[["area", "bedrooms", "age"]],df.price)
```

```
#predict sample value
reg.predict([[4100,6,8]])
```

```
# check coef -> a in ax+b
reg.coef_
```
